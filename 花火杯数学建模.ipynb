```python
import pandas as pd
import numpy as np
import pulp
```

```python
#读取matrix文件
# 使用pandas的read_csv函数读取CSV文件
df = pd.read_csv('df_matrix.csv', encoding='utf-8')

# 打印DataFrame对象，这将显示所有行和列
print(df)
```

```python
#可达矩阵
reachability_matrix=pd.crosstab(df['From'], df['To'])
reachability_matrix
```

```python
#总时效矩阵
Duration_matrix=pd.crosstab(df['From'],df['To'],values=df['Duration'],aggfunc=np.sum)
Duration_matrix
```

```python
# 将 Nan 替换为  0
Duration_matrix[0:20].fillna(0,inplace=True)
Duration_matrix
```

```python
#设施信息表
df_wh=pd.read_csv('df_wh.csv',encoding='utf-8')
df_wh
```

```python
#RDC名称列表
R_wh=df_wh[df_wh['Type']=='RDC']
RDC_list=R_wh['Name'].tolist()
RDC_list
```

```python
#CDC&RDC名称列表
CR_wh=df_wh[df_wh['Type']!='Factory']
CR_list=CR_wh['Name'].tolist()
CR_list
```

```python
#CDC到RDC的运输时效矩阵(CDC可达每个RDC)
c_to_r=Duration_matrix.loc[['nei-lu-CDC','gang-kou-CDC'],RDC_list]
c_to_r
```

```python
r_index=c_to_r.columns.tolist()
r_index.append('nei-lu-CDC')
r_index.append('gang-kou-CDC')
r_index
```

```python
#CDC、RDC到门店的运输时效矩阵
cr_to_st=Duration_matrix.drop(columns=CR_list,axis=1,inplace=False)
cr_to_st.reindex(index=r_index)#行索引与上表列索引保持一致，CDC放在末尾
```

```python
#CDC和RDC可达每个门店,且仓库到自身所在城市耗时为0
for item in CR_list:
    # 选择索引为 item 的行
    row = cr_to_st.loc[item, :]
    
    # 创建一个布尔索引，找出值为 0 的列
    mask = row == 0
    
    # 使用布尔索引找出值为 0 的列的索引
    columns_zero = mask[row == 0].index
    
    # 打印结果
    print(item, '\t', columns_zero, '\n')
```

```python
df_order=pd.read_csv('df_order.csv',encoding='utf-8')
df_order
```

```python
df_order['Name'].value_counts()
```

```python
#需求量矩阵
order_matrix=pd.crosstab(df_order['Name'],df_order['SKU'],values=df_order['qty'],aggfunc=np.sum)
order_matrix
```

```python
#对需求量矩阵行索引进行调整
st_list=cr_to_st.columns.tolist()
order_matrix.reindex(index=st_list)
```

```python
total_or=order_matrix.sum(axis=0)
total_or
```

```python
import pulp
#写问题
SetCoverLP = pulp.LpProblem("SetCover_problem_for_Warehouse", sense=pulp.LpMinimize)
#建立变量x
Zones = list(range(20))
x = pulp.LpVariable.dicts("Zone",Zones,cat="Binary")
Q1_ij=pulp.LpVariable.dict("Q1_ij",(range(2),(range(18))),cat='Continuous')#CDC到RDC的运输量
Q2_jk=pulp.LpVariable.dict("Q2_ij",(range(20),range(104)),cat='Continuous')#CDC、RDC到门店的运输量
area=list(range(20))
z = pulp.LpVariable.dicts('z',area,cat='Binary')
```

```python

#目标函数    
SetCoverLP += pulp.lpSum(pulp.lpSum(0.6*Q1_ij[i][j]*x[j] for i in range(2))for j in range(18))+pulp.lpSum(pulp.lpSum(1.25*Q2_jk[j][k]*x[j]for k in range(104))for j in range(20))+pulp.lpSum(25*x[j]for j in range(20))+pulp.lpSum(pulp.lpSum(6.5*x[j]*(1-0.5*z[j])*Q1_ij[i][j] for j in range(18))for i in range(2))+pulp.lpSum(z[j] for j in range(18))
#约束条件
for i in range(2):
    SetCoverLP += (pulp.lpSum(Q1_ij[i][j] for j in range(20))==total_or[i])
for j in range(20):
    SetCoverLP += (pulp.lpSum(Q2_jk[j][k] for k in range(104))==pulp.lpSum(Q1_ij[i][j] for i in range(2)))
#以RDC为对象分析，RDC出入库相等
for j in range(18):
    for i in range(2):
        SetCoverLP +=(Q1_ij[i][j]==pulp.lpSum(Q2_jk[j][k] for k in range(104)))
#门店需求均被满足
for k in range(104):
    SetCoverLP +=(pulp.lpSum(Q2_jk[j][k]for j in range(20))==order_matrix[k][])

SetCoverLP.solve()
 
print(SetCoverLP.name)
# 格式化输出
temple = "区域 %(zone)d 的决策是：%(status)s"  
# 获得最优解
y = np.array(range(50))
if pulp.LpStatus[SetCoverLP.status] == "Optimal":  
        for i in range(50):
            if x[i].varValue:#将建仓位置存储到数组中
                y[i] = 1
            else:
                y[i] = 0
            output = {'zone': i+1, 'status': '建仓' if x[i].varValue else '--'}
            print(temple % output)
        print("需要建立 {} 个仓库。".format(pulp.value(SetCoverLP.objective)))
```
